```python
import pandas as pd
import numpy as np

def data_thermal_resistance_max(df_mean: pd.DataFrame, df_std: pd.DataFrame, Tmax: float = None):
    """ 
    data_property_max calculates the maximum thermal resistance value of measured thermal properties for the given experiment data.

    use: analysis.data_property_max(df_mean, df_std)
    """

    # Merge the mean and std dataframes on 'Te[K]'
    df = pd.merge(left=df_mean, right=df_std, on=['Te[K]'], suffixes=['_mean', '_std'])
    df['Te[K]'] = df['Te[K]'].round(decimals=0)
    
    # Check if Tmax is provided
    if Tmax is not None:
        if Tmax in df['Te[K]'].values:
            max_row = df[df['Te[K]'] == Tmax]
        else:
            # Find the closest Tmax value in the dataframe
            closest_Tmax = df['Te[K]'].iloc[(df['Te[K]'] - Tmax).abs().argmin()]
            max_row = df[df['Te[K]'] == closest_Tmax]
            print(f"Tmax {Tmax} not found. Using closest value: {closest_Tmax}")
    else:
        # Find the row with the maximum 'TR[K/W]_mean'
        max_row = df[df['TR[K/W]_mean'] == df['TR[K/W]_mean'].max()]
    
    # Extract the values from the selected row
    max_TR_mean = max_row['TR[K/W]_mean'].values[0]
    max_TR_std = max_row['TR[K/W]_std'].values[0]
    max_T = max_row['Te[K]'].values[0]
    max_Q = max_row['Q[W]_mean'].values[0]
    
    # Print the results
    print(f"""
        T_max[K]            : {np.round(max_T, 4)}
        TR[K/W] mean        : {np.round(max_TR_mean, 4)}
        TR[K/W] std         : {np.round(max_TR_std, 4)}
        Q[W]                : {np.round(max_Q, 0)}""")
    
    return max_TR_mean, max_TR_std, max_T
```
