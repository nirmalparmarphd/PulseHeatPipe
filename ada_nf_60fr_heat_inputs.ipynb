```python
# average value of TR
TR_40W_avg = df_mean_40w['TR[K/W]'].mean()
TR_40W_std = df_mean_40w['TR[K/W]'].std()
TR_60W_avg = df_mean_60w['TR[K/W]'].mean()
TR_60W_std = df_mean_60w['TR[K/W]'].std()
TR_80W_avg = df_mean_80w['TR[K/W]'].mean()
TR_80W_std = df_mean_80w['TR[K/W]'].std()
TR_100W_avg = df_mean_100w['TR[K/W]'].mean()
TR_100W_std = df_mean_100w['TR[K/W]'].std()
TR_120W_avg = df_mean_120w['TR[K/W]'].mean()
TR_120W_std = df_mean_120w['TR[K/W]'].std()
```

```python
print(f'Average TR[K/W]: Mean {round(TR_40W_avg,4)} +- Std {round(TR_40W_std,4)}')
print(f'Average TR[K/W]: Mean {round(TR_60W_avg,4)} +- Std {round(TR_60W_std,4)}')
print(f'Average TR[K/W]: Mean {round(TR_80W_avg,4)} +- Std {round(TR_80W_std,4)}')
print(f'Average TR[K/W]: Mean {round(TR_100W_avg,4)} +- Std {round(TR_100W_std,4)}')
print(f'Average TR[K/W]: Mean {round(TR_120W_avg,4)} +- Std {round(TR_120W_std,4)}')
```

```python
# critical value
TR_40W_crt = df_mean_40w[(df_mean_40w['dG[KJ/mol]'] > df_mean_40w['dG[KJ/mol]'].min()) & (df_mean_40w['Te[K]'] == df_mean_40w['Te[K]'].max())]
TR_60W_crt = df_mean_60w[(df_mean_60w['dG[KJ/mol]'] > df_mean_60w['dG[KJ/mol]'].min()) & (df_mean_60w['Te[K]'] == df_mean_60w['Te[K]'].max())]
TR_80W_crt = df_mean_80w[(df_mean_80w['dG[KJ/mol]'] > df_mean_80w['dG[KJ/mol]'].min()) & (df_mean_80w['Te[K]'] == df_mean_80w['Te[K]'].max())]
TR_100W_crt = df_mean_100w[(df_mean_100w['dG[KJ/mol]'] > df_mean_100w['dG[KJ/mol]'].min()) & (df_mean_100w['Te[K]'] == df_mean_100w['Te[K]'].max())]
TR_120W_crt = df_mean_120w[(df_mean_120w['dG[KJ/mol]'] > df_mean_120w['dG[KJ/mol]'].min()) & (df_mean_120w['Te[K]'] == df_mean_120w['Te[K]'].max())]
```
